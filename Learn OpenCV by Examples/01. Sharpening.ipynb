```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
### 1.Sharpening 
### 1.샤프닝 - 커널을 변경함으로써 이미지의 가장자리를 강화하거나 강조하는 효과 구현

image = cv2.imread('../Examples/building.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(image)

# 쉐이핑 커널 생성, 이후 정규화 x
# 행렬의 값은 1로 합함

kernel_sharpening = np.array([[-1,-1,-1],
                              [-1,9,-1],
                              [-1,-1,-1]])

# 입력 이미지에 다른 커널 적용
sharpened = cv2.filter2D(image, -1, kernel_sharpening)

plt.subplot(1,2,2)
plt.title('Image Sharpening')
plt.imshow(sharpened)

plt.show()
```
